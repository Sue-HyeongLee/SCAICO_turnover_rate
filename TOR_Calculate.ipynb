```python
import numpy as np
import pandas as pd
```

# **DATA IMPORT**

```python
tp_adv = pd.read_csv("input/topic_modeling_result/bank_adv_topic_modeling.csv")
tp_dadv = pd.read_csv("input/topic_modeling_result/bank_dadv_topic_modeling.csv")

fv = pd.read_csv("input/financial_variable/bank_financial_financial_variable.csv")
tor = pd.read_csv("input/turn_over_rate/bank_financial_business_turn_over_rate.csv")
```

```python
print(len(tp_adv))
print(len(tp_dadv))
print(len(fv))
print(len(tor))
```

```python
tp_adv = tp_adv.drop(['Unnamed: 0'], axis=1)
tp_dadv = tp_dadv.drop(['Unnamed: 0'], axis=1)
fv = fv.drop(['Unnamed: 0'], axis=1)
tor = tor.drop(['Unnamed: 0'], axis=1)
```

```python
# Preprocessing for topic modeling result
cn_adv = tp_adv['cmp']
cn_dadv = tp_dadv['cmp']

for i, val in enumerate(cn_adv):
  if '(주)' in val:
    tmp = val.replace('(주)','')
    tp_adv.loc[i,'cmp'] = tmp

for i, val in enumerate(cn_dadv):
  if '(주)' in val:
    tmp = val.replace('(주)','')
    tp_dadv.loc[i,'cmp'] = tmp

col_adv_list = list(tp_adv.columns)
col_adv_list[0] = 'company_name'
for i in range(1, len(tp_adv.columns)):
  col_adv_list[i] = 'adv_topic ' + str(col_adv_list[i])
tp_adv.columns = col_adv_list

col_dadv_list = list(tp_dadv.columns)
col_dadv_list[0] = 'company_name'
for i in range(1, len(tp_dadv.columns)):
  col_dadv_list[i] = 'dadv_topic ' + str(col_dadv_list[i])
tp_dadv.columns = col_dadv_list

tp_dadv
```

```python
# Preprocessing for financial variable

afv = fv['average_salary']
tfv = fv['total_sale']

for i, val in enumerate(afv):
  if '만원' in val:
    tmp = int(val[:-2].replace(',',''))
    fv.loc[i,'average_salary'] = tmp
  if '회사' in val:
    fv.loc[i,'average_salary'] = np.NAN


for i, val in enumerate(tfv):
  if '조원' in val:
    tmp = float(val[:-2].replace(',',''))*1000
    fv.loc[i,'total_sale'] = tmp
  if '억원' in val:
    tmp = float(val[:-2].replace(',',''))
    fv.loc[i,'total_sale'] = tmp
  if '회사' in val:
    fv.loc[i,'total_sale'] = np.NAN

fv
```

```python
# Preprocessing for turn over rate

tor_val = tor['turn_over_rate']

for i, val in enumerate(tor_val):
  if '정보' in val:
    tor.loc[i,'turn_over_rate'] = np.NAN
  elif val == '(9999%)' :
    tor.loc[i,'turn_over_rate'] = np.NAN
  elif '(' in val:
    tmp = float(val[1:-1].replace('%',''))*0.01
    tor.loc[i,'turn_over_rate'] = tmp

tor
```

```python
df = pd.merge(tp_adv, fv, on = 'company_name', how = 'inner')
df = pd.merge(df, tp_dadv, on = 'company_name', how = 'inner')
df = pd.merge(df, tor, on = 'company_name', how = 'inner')
df = df.dropna()
df
```

# **MACHINE LEARNING MODEL**

```python
# Linear Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

y_target = df['turn_over_rate']
x_data = df.drop(['company_name','turn_over_rate'], axis = 1, inplace = False)

X_train, X_test, y_train, y_test = train_test_split(x_data, y_target, test_size = 0.3, random_state = 7)

lr = LinearRegression()
lr.fit(X_train, y_train)

y_preds = lr.predict(X_test)
mse = mean_squared_error(y_test, y_preds)
rmse = np.sqrt(mse)

print('MSE: {0: .3f}, RMSE : {1: .3F}'.format(mse, rmse))
print('Variance score : {0:.3f}'.format(r2_score(y_test,y_preds)))
```
