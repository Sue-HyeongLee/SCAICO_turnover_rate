```python
# Preprocessing for turn over rate
def tor_preprocessing(tor):
  tor_df = tor
  tor_val = tor_df['turn_over_rate']

  for i, val in enumerate(tor_val):
    if '정보' in val:
      tor_df.loc[i,'turn_over_rate'] = np.NAN
    elif val == '(9999%)' :
      tor_df.loc[i,'turn_over_rate'] = np.NAN
    elif '(' in val:
      tmp = float(val[1:-1].replace('%',''))*0.01
      tor_df.loc[i,'turn_over_rate'] = tmp

  return tor_df
```

```python
# Preprocessing for financial variable
def fv_preprocessing(fv):
  fv_df = fv
  afv = fv_df['average_salary']
  tfv = fv_df['total_sale']

  for i, val in enumerate(afv):
    if type(val) == str:
        if '만원' in val:
          tmp = int(val[:-2].replace(',',''))
          fv_df.loc[i,'average_salary'] = tmp
        if '회사' in val:
          fv_df.loc[i,'average_salary'] = np.NAN
        if '수집' in val:
          fv_df.loc[i,'average_salary'] = np.NAN

  for i, val in enumerate(tfv):
    if type(val) == str:
        if '조원' in val:
          tmp = float(val[:-2].replace(',',''))*1000
          fv_df.loc[i,'total_sale'] = tmp
        if '억원' in val:
          tmp = float(val[:-2].replace(',',''))
          fv_df.loc[i,'total_sale'] = tmp
        if '억' in val:
          tmp = float(val[:-1].replace(',',''))
          fv_df.loc[i,'total_sale'] = tmp
          print(fv_df.loc[i,'total_sale'])
          print(i)
        if '회사' in val:
          fv_df.loc[i,'total_sale'] = np.NAN

  return fv_df
```

```python
from sklearn.model_selection import GridSearchCV

def print_best_params(model, params):
    grid_model = GridSearchCV(model, param_grid=params,
                              scoring='r2', cv=5)
    grid_model.fit(X_train,y_train)
    r2_score = grid_model.best_score_
    print('train: {0} After 5 CV, best average R2 score: {1}, best alpha: {2}'.format(model.__class__.__name__,np.round(r2_score,4),grid_model.best_params_))
    print('test: {0} After 5 CV, best average R2 score: {1}, best alpha: {2}'.format(model.__class__.__name__,np.round(grid_model.score(X_test, y_test),4),grid_model.best_params_))
    grid_model = GridSearchCV(model, param_grid=params,
                              scoring='neg_mean_squared_error', cv=5)
    grid_model.fit(X_train,y_train)
    mse = -1 * grid_model.best_score_
    rmse = np.sqrt(-1*grid_model.best_score_)
    print('train: {0} After 5 CV, best average MSE: {1}, best average RMSE: {2}, best alpha: {3}'.format(model.__class__.__name__,np.round(mse,4),
                                                                           np.round(rmse,4),grid_model.best_params_))

    grid_score = grid_model.score(X_test, y_test)
    mse = -1 * grid_score
    rmse = np.sqrt(-1*grid_score)
    print('test: {0} After 5 CV, best average MSE: {1}, best average RMSE: {2}, best alpha: {3}'.format(model.__class__.__name__,np.round(mse,4),
                                                                           np.round(rmse,4),grid_model.best_params_))

    grid_model = GridSearchCV(model, param_grid=params,
                              scoring='neg_mean_absolute_error', cv=5)
    grid_model.fit(X_train,y_train)
    mae = -1 * grid_model.best_score_
    print('train: {0} After 5 CV, best average MAE: {1}, best alpha: {2}'.format(model.__class__.__name__,np.round(mae,4),grid_model.best_params_))
    print('test: {0} After 5 CV, best average MAE: {1}, best alpha: {2}'.format(model.__class__.__name__,np.round(-1*grid_model.score(X_test, y_test),4), grid_model.best_params_))




ridge_params = {'alpha':[ 1, 2, 3, 4, 5]}

print_best_params(ridge_reg,ridge_params)
```
