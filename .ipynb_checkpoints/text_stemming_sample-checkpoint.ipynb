```python
# sample review text
# 로픽원테크놀로지스(유)

const_doc = "주 4.5일제, 매주 금요일 1시 퇴근. 자유로운 연차 사용. 무제한 연차. 생일자 유급 휴가. 생일자는 출근하지 않는다. 시스템이 많이 갖춰져 있고, 선도적인 기술 시스템을 많이 검토하고 기업에 알맞게 도입 적용하며, 도입 시, 충분히 설명하고 공감대를 형성하며 도입을 진행한다. - PLS(Project Life Streaming, PMS, 프로젝트 관리 시스템): 프로젝트 진행 현황이 모두에게 공개됨. 공동체의 성장을 위함. - Git(코드 형상관리) - G Suite(Google 업무 협업 시스템): 구글 클라우드 문서 기반 시스템에서 문서 생성. 협업. - One Source, Multi-Use를 항상 염두에 두고, 업무 고 효율화를 추구. 어떤 업무, 미션을 자기 주도적으로 문제 해결을 하면서 나아가는 사람에게는 유리. - 자기 주도적으로 하면서, 기업 자산화를 해가면서 나아가는 시스템. - 문제 해결에 초점. 고객에게 가치 있는 서비스 창출 중심. 업무 자율도 높음. 기본적으로 속도도 빠르고 강도도 높고 업무 센스도 높아야 하기에 스스로 의욕적이고 문제 해결에 자신이 있다면, 충분히 재밌는 곳이 될 것으로 생각. 스스로 문제, 이슈, 태스크, 업무를 정의할 수 있고, 자신이 정의해서 업무를 진행하기에 업무 자율도가 매우 높음."
# 해당 텍스트는 로픽원테크놀로지스의 장점 리뷰 중 한가지만을 가져온 텍스트 데이터이다.
# 프로젝트에서는 한 기업에 있는 장점 리뷰들을 하나로 모은 텍스트 파일 하나, 단점 리뷰들을 하나로 모은 텍스트 파일 하나, 총 두가지 파일을 활용해서 stemming을 진행한다.
# 결과물은 stemming 된 장점 딕셔너리와 단점 딕셔너리 두가지로 나온다.
```

```python
print(type(const_doc))
print(len(const_doc))
print(const_doc[:600])
```

```python
from konlpy.tag import Okt

t = Okt()
tokens_const = t.morphs(const_doc)

print('# tokens:',len(tokens_const))
print('# before 100th token')
print(tokens_const[:100])
```

```python
tokens_const = t.nouns(const_doc)
print('# noun tokens:',len(tokens_const))
print('# before 100th token')
print(tokens_const[:100])
```

```python
tokens_const = [token for token in tokens_const if len(token) >1]
print('# tokens:', len(tokens_const))
print('# before 100th token')
print(tokens_const[:100])
```

```python
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt

font_name = 'AppleGothic'
rc('font', family = font_name)

const_cnt = {}

for word in tokens_const:
    const_cnt[word] = const_cnt.get(word, 0)+1

def word_graph(cnt, max_words = 10):
    sorted_w = sorted(cnt.items(), key=lambda kv: kv[1])
    print(sorted_w[-max_words:])
    n, w = zip(*sorted_w[-max_words:])

    plt.barh(range(len(n)),w, tick_label=n)
    plt.show()

word_graph(const_cnt, max_words=30)
```
